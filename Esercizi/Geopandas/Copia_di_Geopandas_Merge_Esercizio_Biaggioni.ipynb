```python
provincie = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/ProvCM01012021_g')
provincie
```

```python
regioni = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Reg01012021_g')
regioni
```

```python
comuni = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Com01012021_g')
comuni
```

visualizzare il nome della provincia e il nome della regione

```python
pd.merge(provincie,regioni,how='inner',on='COD_REG')
```

visualizzare il comune e in quale regione si trova

```python
pd.merge(comuni,regioni,how='inner',on='COD_REG')
```

numero di comuni per ogni regione

```python
pd.merge(comuni,regioni,how='inner',on='COD_REG').groupby('DEN_REG')['COMUNE'].count().sort_values()
```

numero di comuni per ogni provincia

```python
pd.merge(comuni,provincie,how='inner',on='COD_PROV').groupby('DEN_UTS')['COMUNE'].count().sort_values()
```

numero di provincie per ogni regione

```python
pd.merge(provincie,regioni,how='inner',on='COD_REG').groupby('DEN_REG')['DEN_UTS'].count().sort_values()
```
