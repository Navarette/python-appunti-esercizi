```python
province =  geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/ProvCM01012021_g')
province
```

```python
regioni = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Reg01012021_g')
regioni
```

```python
provLomb = province[province.within(regioni[regioni.DEN_REG.str.contains('Lombardia')].geometry.squeeze())]
ax = autostrade.to_crs(epsg=3857).plot(color='grey')
provLomb.to_crs(epsg=3857).plot(facecolor='none',edgecolor='K',ax=ax)
ctx.add_basemap(ax)
```

2.

```python
userRoad = input()
if userRoad not in autostrade.COD_PE.tolist():
  print('cicuccia')
else:
  auto = autostrade[autostrade.COD_PE == userRoad]
  autoProv = provLomb[provLomb.intersects(auto.unary_union)]
  ax = auto.to_crs(epsg=3857).plot()
  autoProv.to_crs(epsg=3857).plot(ax=ax,facecolor='lime',alpha=.5)
  ctx.add_basemap(ax)
```

3.

```python
comuni = comuni[comuni.within(regioni[regioni.DEN_REG.str.contains('Lombardia')].geometry.squeeze())]
userRoad = input()
distance = float(input())

ax = comuni[comuni.distance(autostrade[autostrade.COD_PE == userRoad].unary_union) <= distance].to_crs(epsg=3857).plot(facecolor='r')
autostrade[autostrade.COD_PE == userRoad].to_crs(epsg=3857).plot(color='k',ax=ax)
ctx.add_basemap(ax)
```
